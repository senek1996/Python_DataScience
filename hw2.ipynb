```python
#задание 2
t=np.linspace(0,10,51)
print(t)
```

```python
plt.plot(t,f,'g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5,9.5,-2.5, 2.5])
```

```python
#задание 3
x=np.linspace(-3,3,51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
from matplotlib import rcParams
fig, axs = plt.subplots(2,2)
rcParams["figure.figsize"] = 8, 6

axs[0,0].plot(x,y1)
axs[0,0].set_title('График y1')
axs[0,0].set_ylabel('y1')
axs[0,0].set_xlim([-5,5])

axs[0,1].plot(x,y2)
axs[0,1].set_title('График y2')
axs[0,1].set_ylabel('y2')

axs[1,0].plot(x,y3)
axs[1,0].set_title('График y3')
axs[1,0].set_ylabel('y3')

axs[1,1].plot(x,y4)
axs[1,1].set_title('График y4')
axs[1,1].set_ylabel('y4')

#непонятно, как делать зазоры. Может так?
for i in range(2):
    for j in range(2):
        axs[i,j].set_aspect(0.3)
```

```python
#задание 4
import pandas as pd

plt.style.use('fivethirtyeight')
dataset=pd.read_csv('creditcard.csv')
```

```python
dataset.columns
```

```python
classes_cou=dataset['Class'].value_counts()
classes_cou
classes_cou.plot(kind='bar')
plt.axis([0,1,0,300000])
```

```python
classes_cou.plot(kind='bar', logy=True)
```

```python
v1_usual=dataset[dataset['Class']==0]['V1']
v1_fraud=dataset[dataset['Class']==1]['V1']
print(v1_usual.size,' ',v1_fraud.size,' ',dataset['Class'].size)
```

```python
plt.hist(v1_usual,bins=20,alpha=0.5,density=True,color='gray')
plt.hist(v1_fraud,bins=20,alpha=0.5,density=True,color='red')
plt.legend(('Class 0','Class 1'))
plt.xlabel('V1')
```
